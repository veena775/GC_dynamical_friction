```python
import numpy as np
import matplotlib as mpl
from matplotlib import cm
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import curve_fit
from scipy import asarray as ar,exp
from scipy.stats import norm
import math
import statistics
```

```python
###import data
dataGC = np.genfromtxt('best_sample.txt', skip_header=1, delimiter=',') ## from Shany's paper
M_GC = dataGC[:, 2] # log GC Mass (in solar masses??)
M_S = dataGC[:, 4] #log Stellar Mass


###plot a scatter plot of said data
plt.rc('text', usetex=True)
plt.rc('font', family='serif',size=15)
plt.scatter(M_S, M_GC/M_S, s=5, color = 'black')  ## Mgc or Mgc/Ms????
plt.xlabel('log stellar mass', fontsize=15)
plt.ylabel('log GC mass / log stellar mass', fontsize=15)
plt.show()

plt.scatter(M_S, M_GC, s=5, color = 'black')  ## Mgc or Mgc/Ms????
plt.xlabel('log stellar mass', fontsize=15)
plt.ylabel('log GC mass', fontsize=15)
plt.show()


######### VERTICAL STRIPE

### choose a range of stellar masses eg 8-8.25
M_GCstripe = [] 
M_Sstripe = []
M_GCstripe_ratio = []
n = len(M_GC)

## 7 - 8.25
## 8.25 - 9
## or nonlinear bins
lbound = 8.5     #lower bound of stellar mass of stripe
ubound = 9  #upper bound of stellar mass of stripe

## get stellar and GC masses within stripe 
for i in range (n):
    if M_S[i] >= lbound and M_S[i] <= ubound and M_GC[i]!=0:
        M_GCstripe.append(M_GC[i])
        M_Sstripe.append(M_S[i])
        M_GCstripe_ratio.append(M_GC[i]/M_S[i])     
        
plt.scatter(M_Sstripe, M_GCstripe, s=5, color='black')
plt.xlabel('log stellar mass', fontsize=15)
plt.ylabel('log GC mass', fontsize=15)
plt.show()


##########  HISTOGRAM of counts at certain Stellar mass
# # plots M GCs / M stellar counts for specified vertical stripe
# plt.hist(MGCstripe, bins = 10)
# plt.xlabel('log GC Mass')
# plt.xticks(np.arange(min(MGCstripe), max(MGCstripe)+0.1, 0.5))
# plt.ylabel('counts')
# plt.show

#########   DISTRIBUTION

def gaussian(x, mu, sigma):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))


def Gaussian_bincenters(M_GCstripe):
    counts, bin_edges = np.histogram(M_GCstripe, bins=10, density=True)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    return [counts, bin_centers]

def Gaussian_params(M_GCstripe):   
    p0 = [np.mean(M_GCstripe), np.std(M_GCstripe)]  # Initial guess for the parameters
    params, cov = curve_fit(gaussian, Gaussian_bincenters(M_GCstripe)[1], Gaussian_bincenters(M_GCstripe)[0], p0=p0)
    return params

params = Gaussian_params(M_GCstripe)
print(params)
print()
x_values = np.linspace(4.5, 7, 100)
# counts, bin_edges = np.histogram(M_GCstripe, bins=10)
# bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
# print(bin_centers)
# p0 = [np.mean(M_GCstripe), np.std(M_GCstripe)]
# print(p0)
# params, cov = curve_fit(gaussian, bin_centers, counts, p0=p0)
# print(params)
# print(M_GCstripe)

plt.hist(M_GCstripe, bins=10, alpha=0.6, label='GC Mass Counts', color='coral', edgecolor = 'black', density = True)  #density=True
#plt.plot(Gaussian_bincenters(M_GCstripe), gaussian(Gaussian_bincenters(M_GCstripe), *Gaussian_params(M_GCstripe)), 'r-', label='Gaussian Fit')


plt.plot(x_values, gaussian(x_values, *params), 'r-', label='Gaussian Fit')



plt.xlabel('Log GC Mass')
plt.ylabel('Counts')
plt.legend()
plt.show()

####### SAMPLING MASSES FROM DISTRIBUTION
##sampling GC Mass for a given Stellar Mass
mu, sigma = params
sample_masses = np.random.normal(mu, sigma, 5000)
count, bins, ignoored = plt.hist(sample_masses, 50, density=True, color = 'coral', alpha=0.8)
plt.plot(bins, gaussian(bins, mu, sigma), linewidth=2, color='r')
plt.show()
```
