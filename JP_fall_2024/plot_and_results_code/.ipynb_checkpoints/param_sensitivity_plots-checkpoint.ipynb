```python
# grid_data = np.load("updated_grid_merger_classification.npy")
# R_es, GC_masses, halo_masses, mergers = grid_data[:, 0], grid_data[:, 1], grid_data[:, 2], grid_data[:, 3]


trial_num = 'uniform_final_v1'
param_values = np.load('param_values/param_values_trial'+str(trial_num)+'.npy')
R_es = param_values[:,0]  #R_es in pc
GC_masses = param_values[:,1]  #total GC mass in log Solar Mass
halo_masses = param_values[:,2]  #DM halo in log Solar Mass

GC_masses = np.append(GC_masses, [10**5.5, 10**5.7, 10**5.8, 
                                  10**5.9, 10**6.1, 10**6.2, 10**6.3, 
                                  10**6.4, 10**6.6, 10**6.7,
                                 10**6.7, 10**6.8, 10**6.9, 10**7.1, 
                                  10**7.2, 10**7.3, 10**7.4])
R_es = np.append(R_es, [10**3.5, 10**2.9, 10**2.8, 10**3.1, 10**3.3])

value_to_remove = 10**8
GC_masses = GC_masses[GC_masses != value_to_remove]
value_to_remove = 10**5
GC_masses = GC_masses[GC_masses != value_to_remove]
```

### Re

```python
NSC_formation_time = []
NSC_formation_time_std = []
NSC_formation_mass = []
NSC_formation_mass_std = []
formation_Re = []
max_masses = []

avg_mass_over_n = []
avg_time_over_n = []
avg_mass_std = []
avg_time_std = []


max_masses = []
Re_for_mm =[]

path_name = '/home/vk9342/JP_fall_2024/test_sim_run_trial_'+str(trial_num)+'/'
trial_num = 'uniform_final_v1'

        
for R_e in R_es:
    NSC_times_for_Re = []
    NSC_masses_for_Re = []
    max_masses_for_Re = []

    for GC_mass in GC_masses:
        for halo_mass in halo_masses:

            NSC_masses = np.zeros([10, 21])
        
            R_e_sim = np.round(R_e / 1000, 3)
            GC_mass_sim = np.round(GC_mass * 1e-5, 3)
            halo_mass_sim = np.round(halo_mass, 1)
            file_name20 = 'Sim_NFW_20_'+str(R_e_sim)+'_'+str(GC_mass_sim)+'_'+str(halo_mass_sim)+'.txt'  #final
            file20 = os.path.join(path_name, file_name20)
            if os.path.exists(file20):
                for output_number in range(21): #loop through all files
                    file_name = 'Sim_NFW_'+str(output_number)+'_'+str(R_e_sim)+'_'+str(GC_mass_sim)+'_'+str(halo_mass_sim)+'.txt' 
                    file = os.path.join(path_name, file_name)
                    # print(file_name)
                    data = np.genfromtxt(file, skip_header=2)
                    iterations = data[:, 0]
                    masses = data[:, 1]
                    for k in range(10): #loop through all iterations in file
                        relevant_masses = masses[iterations==k]
                        NSC_masses[k, output_number] = max(relevant_masses)
                avg_NSC_masses = np.average(NSC_masses, axis=0)  #avg nsc mass for each timestep
                max_mass = max(avg_NSC_masses)
            
                final_mass = (avg_NSC_masses[20])

                i = 0
                while i < 21:
                    if avg_NSC_masses[i] / max_mass >= 0.9 and max_mass >= 20:
                        NSC_times_for_Re.append(i / 2)
                        NSC_masses_for_Re.append(avg_NSC_masses[i])
                        max_masses_for_Re.append(max_mass)
                        break
                    i += 1
            # else:
            #     print(f'File {file_name20} not found')

    # Average over all GC and halo mass combinations for the current R_e
    if NSC_times_for_Re:
        NSC_formation_time.append(np.mean(NSC_times_for_Re))
        NSC_formation_time_std.append(np.std(NSC_times_for_Re))
        NSC_formation_mass.append(np.mean(NSC_masses_for_Re))
        NSC_formation_mass_std.append(np.std(NSC_masses_for_Re))
        formation_Re.append(R_e)
        max_masses.append(np.mean(max_masses_for_Re))
            
```
